```python
import pandas as pd
import numpy as np
import scipy
from scipy import linalg as LA
from numpy import linalg as la
from sklearn.decomposition import PCA
from sklearn import linear_model

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data_norm = data.sub(data.mean(0), axis=1).div(data.std(0), axis=1)
sns.pairplot(data_norm, hue='quality', palette=sns.cubehelix_palette(6))
```

```python
data_X = data.drop('quality', axis=1)
data_Y = data['quality']
X = data_X.as_matrix()
Y = data_Y.as_matrix()
```

```python
# clf = linear_model.LinearRegression(normalize=True)
clf = linear_model.LinearRegression()
X_norm = (X - X.mean(axis=0)) / X.std(axis=0)
Y_norm = (Y - Y.mean(axis=0)) / Y.std(axis=0)
clf.fit(X_norm, Y_norm)
```

```python
print(pd.DataFrame({'Name':data_X.columns, 'Coef': np.abs(clf.coef_)}).sort_values(by='Coef'))
```

```python
print(pd.DataFrame({'Name':data_X.columns, 'Coef': clf.coef_}))
```
