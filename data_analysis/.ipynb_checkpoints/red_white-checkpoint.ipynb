```python
import pandas as pd
import numpy as np
import scipy
from scipy import linalg as LA
from numpy import linalg as la
from sklearn.decomposition import PCA
from sklearn import linear_model

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# headers = ('class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols')
# headers = headers + ('Proanthocyanins', 'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline')
# print(headers)
# data = pd.read_csv('wine.data', names=headers)
val_num = 100

data_red = pd.read_csv('winequality-red.csv', sep=';')
data_white = pd.read_csv('winequality-white.csv', sep=';')
data_red_val = data_red[-val_num:]
data_white_val = data_white[-val_num:]
data_red = data_red[:-val_num]
data_white = data_white[:-val_num]
data_red['type'] = np.repeat('red', len(data_red))
data_white['type'] = np.repeat('white', len(data_white))
data = data_red.append(data_white, ignore_index=True)
data_val = data_red_val.append(data_white_val, ignore_index=True)
data
```

```python
data_norm = data[list(range(12))].sub(data[list(range(12))].mean(0), axis=1).div(data[list(range(12))].std(0), axis=1)
# sns.pairplot(data_norm, vars=['alcohol', 'quality'], hue='type')
data_norm['type'] = data['type']
# sns.pairplot(data_norm, hue='type', palette=sns.cubehelix_palette(2))
```

```python
data_X = data.drop('type', axis=1)
data_X = data_X.drop('quality', axis=1)

data_X_val = data_val.drop('quality', axis=1)

Y = np.r_[np.repeat(1., len(data_red)), np.repeat(-1., len(data_white))]
Y_val = np.r_[np.repeat(1., len(data_red_val)), np.repeat(-1., len(data_white_val))]
X = data_X.as_matrix()
X_val = data_X_val.as_matrix()
# Y = data_Y.as_matrix()
```

```python
# clf = linear_model.LinearRegression(normalize=True)
clf = linear_model.LinearRegression()
X_norm = (X - X.mean(axis=0)) / X.std(axis=0)
X_val_norm = (X_val - X_val.mean(axis=0)) / X_val.std(axis=0)
clf.fit(X_norm, Y)
```

```python
coef_df = pd.DataFrame({'Name':data_X.columns, 'Coefficients': np.abs(clf.coef_)}).sort_values(by='Coefficients')
print(coef_df)
# plt.rcParams["figure.dpi"] = 1200
plt.rcParams["savefig.dpi"] = 400
ax = sns.barplot(data=coef_df, x='Name', y='Coefficients')
ax.set_xticklabels(coef_df['Name'], rotation=70)
ax.set(xlabel='Attribute', ylabel='abs(Coefficient)')
plt.savefig('coef_graph')
```

```python
print(pd.DataFrame({'Name':data_X.columns, 'Coef': clf.coef_}))
```

```python
Y_predict = clf.predict(X_norm)
correct_count = 0
for i in range(len(Y_predict)):
    result = Y_predict[i] * Y[i]
    if result > 0:
        correct_count += 1

print(correct_count / len(Y_predict))
```

```python
Y_val_predict = clf.predict(X_val_norm)
cross_agg = np.zeros((2, 2))
correct_count = 0
for i in range(len(Y_val_predict)):
    k1 = 0 if Y_val_predict[i] > 0 else 1
    k2 = 0 if Y_val[i] > 0 else 1
    cross_agg[k1, k2] += 1
    result = Y_val_predict[i] * Y_val[i]
    if result > 0:
        correct_count += 1

print(correct_count / len(Y_val_predict))
```
